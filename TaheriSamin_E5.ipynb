**Exercise 9.5: Singular values**

```python
import numpy
import random
from scipy.linalg import toeplitz
import scipy

n = int(input('input n: '))
C = []
p = float(input('input P: '))
for i in range(0, n):
    temp = []
    for j in range(0, n):
        if random.uniform(0, 1) > p:
            temp.append(0)
        else:
            temp.append(1)
    C.append(temp)

U, sigma, VT = scipy.linalg.svd(C)
print(sigma)
```

**Exercise 10.1: Least squares**

```python
import numpy as np
import scipy.optimize as opt
 
m, n = 100, 50
A = np.random.rand(m, n) # Generate matrix A ∈ Rm × n with m > n
b = np.random.random(m) # generate a vector b ∈ Rm.
res = opt.lsq_linear(A, b) # find x = argminx k Ax − bk 2.
x = res.x
residual = b - np.dot(A, x) # Print the norm of the residual.
norm = np.linalg.norm(residual) #Frobenius norm
print(norm)
```

**Exercise 10.2: Optimization**

```python
import numpy as np
import scipy.optimize as opt
 
def f(x):  
	res = np.power(np.sin(x-2), 2) * np.exp(-1*(x**2))
	return -1 * res # Change the function into opposite one
    
res = opt.minimize_scalar(f) # Find the maximum of the function
x = res.x
print(-1*f(x)) # Equivalent to find the minimum of the opposite function
```

**Exercise 10.3: Pairwise distances**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import scipy.spatial.distance as dis
 
n, m = 50, 100
X = np.random.rand(n, m)
d = dis.pdist(X)
 
# As an example application, consider n cities, and we are given their coordinates in two columns
x = []
y = []
for a in range(0, n//2):
	x = x + [a for _ in range(1, n)]
	y = y + [b for b in range(1, n)]
 
x = np.array(x) # set x lable
y = np.array(y) # set y lable
 
# Now we want a nice table that tells us for each two cities, how far they are apart.
figure = plt.figure()
ax = Axes3D(figure)
plt.title("Distance between two cities")  
ax.set_xlabel ('Firt City')
ax.set_ylabel('Second City')
ax.set_zlabel('Distance')
ax.scatter(x, y, d, c='r')
plt.show()
```

**11 Matplotlib**

**Exercise 11.1: Plotting a function**

```python
import matplotlib.pyplot as plt
import numpy as np

size = 100
f, ax = plt.subplots()
ones = np.ones(size)
x = np.linspace(0., 2., size)
y1 = 0.5 * (ones - np.cos(2 * (x - ones * 2))) # sin^2(x - 2) = 0.5 * (1 - cos2(x - 2))
y2 = np.power(np.e, -1 * np.multiply(x, x)) # e^(-x^2)
y = np.multiply(y1, y2) # y1 * y2 mathematically, not the same as matrix dot product
ax.plot(x, y)
ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title("$sin^2(x-2){e^{-x^2}}$")
plt.tight_layout()
plt.show()
```

**Exercise 11.2: Data**

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
X = np.random.random_sample((20, 10)) * 10
b = np.random.random(10) * 3 - 1.5
z = np.random.normal((20,))
y = X.dot(b) + z
b_ = np.array(np.linalg.lstsq(X, y, rcond=-1)[0])
x = np.arange(0, 10)

f, ax = plt.subplots()
ax.set_xlim(0, 9)
ax.set_ylim(-1.5, 1.5)
ax.set_xlabel("index")
ax.set_ylabel("value")
ax.plot(x, b, 'rx', label='True coefficients')
ax.plot(x, b_, 'bo', label='Estimated coefficients')
plt.hlines(0, 0, 9, colors='k', linestyle="--")
plt.tight_layout()
plt.show()
```
